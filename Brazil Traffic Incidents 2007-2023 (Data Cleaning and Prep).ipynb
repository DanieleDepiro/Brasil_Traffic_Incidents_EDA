```python
wrong_values = []
for date in mdf['data_inversa']:
    if len(str(date)) < 10: # runned for less and more
        wrong_values.append(date)
        
for value in wrong_values:
    print(value)
```

```python
mdf['data_inversa'] = mdf['data_inversa'].str.replace(r'/16$', '/2016')
```

```python
# Converting 'yyyy-mm-dd' format to 'dd/mm/yyyy' format
mdf['data_inversa'] = mdf['data_inversa'].str.replace(r'(\d{4})-(\d{2})-(\d{2})', r'\3/\2/\1', regex=True)

# Ensure all dates are in desired format
mdf['data_inversa'] = pd.to_datetime(mdf['data_inversa'], format='%d/%m/%Y').dt.strftime('%d/%m/%Y')
```

```python
# final check all are in the right format
converted_dates = pd.to_datetime(mdf['data_inversa'], format='%d/%m/%Y', errors='coerce')

all_values_in_correct_format = converted_dates.notna().all()

print(all_values_in_correct_format)
```
